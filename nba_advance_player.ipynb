```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
nba_advanced = pd.read_csv('Resources/NBA Season Data.csv')
nba_adv = nba_advanced[['Player', 'Tm', 'PER', 'BPM', 'VORP', 'WS', 'Team Mar', 'Year']]
nba_adv2 = nba_advanced[nba_advanced['Year'] > 1985]
nba_adv2 = nba_advanced[nba_advanced['MP'] > 3300]
nba_adv2.count()
```

```python
per_corr = nba_adv2['Team Mar'].corr(nba_adv2['PER'])
bpm_corr = nba_adv2['Team Mar'].corr(nba_adv2['BPM'])
vorp_corr = nba_adv2['Team Mar'].corr(nba_adv2['VORP'])
ws_corr = nba_adv2['Team Mar'].corr(nba_adv2['WS'])
```

```python
vc_slope, vc_int, vc_r, vc_p, vc_std_err = st.linregress(nba_adv2['Team Mar'], nba_adv2['PER']) 
vc_fit = vc_slope * nba_adv2['Team Mar'] + vc_int 
equation = 'y= ' + str(round(vc_slope, 2)) + 'x + ' + str(round(vc_int, 2))


plt.scatter(nba_adv2['Team Mar'], nba_adv2['PER'])
plt.plot(nba_adv2['Team Mar'],vc_fit,"--", color = 'green')

plt.xlabel('Team Margin')
plt.ylabel('PER')
plt.title('PER Graph',fontsize =15)
plt.annotate(equation, xy=(2, 30), fontsize=10, color='orange')

plt.savefig("output_images/PER Graph")
```
